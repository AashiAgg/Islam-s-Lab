```python
# Using RegularSeasonCompactResults.csv
# 1- Create a filter to show if there are any draw matches
# 2- Apply the filter to the dataframe
# 3- SHOW the dataframe length or count (4 marks)
## TODO: Your code should be here


df = pd.read_csv('RegularSeasonCompactResults.csv')

def is_draw(row):
    return row['Wscore'] == row['Lscore']


draw_matches = df[df.apply(is_draw, axis=1)]


draw_match_count = len(draw_matches)
print("Number of draw matches:", draw_match_count)



```

```python
# Using RegularSeasonCompactResults.csv
# 1- Create a filter showing the entries where the winningteamID= 1400- (last two digits of your id)
# 2- Create another filter for the season of 2013
# 3- Apply the filter to the dataframe
# 4- Show the resulted dataframe (4 marks)
# (Example: IF your id is 8000021, the winning team id = 1400-21= 1389)
# TODO: Your code should be here

# Load the dataset
df = pd.read_csv('RegularSeasonCompactResults.csv')

# Create a filter showing the entries where the winningteamID= 1400- (last two digits of your id)

target_team_id = 1400 - 99

# filter for winning team ID
winning_team_filter = df['Wteam'] == target_team_id

# filter for the season of 2013
season_filter = df['Season'] == 2013

# Applying filters to the DataFrame
filtered_df = df[winning_team_filter & season_filter]
print(filtered_df)
```

```python
# Using RegularSeasonCompactResults.csv & Teams.csv
# 1- Load and merge both dataframes
# 2- Where the merge is on WinningTeamId (RegularSeasonCompactResults) and TeamId (Teams.csv)
# 3- Name the dataframe: DFCombined
# 4- Show the top 3 rows (4 marks)
# TODO: Your code should be here

regular_season_df = pd.read_csv('RegularSeasonCompactResults.csv')
teams_df = pd.read_csv('Teams.csv')

# Merging
DFCombined = regular_season_df.merge(teams_df, left_on='Wteam', right_on='Team_Id', how='inner')

# Show the top 3 rows of the combined DataFrame
top_3_rows = DFCombined.head(3)
print(top_3_rows)

```

**1. .** (5 marks)The relationship between the penguin's bill length and bill depth for each species

```python
## TODO: Check your attributes, make a graph using Matplotlib or Seaborn to show the relationship needed
#pyplot  bargraph
#bargraph best for relationship fits
import pandas as pd
import matplotlib.pyplot as plt

# Load the penguins.csv file into a DataFrame
df = pd.read_csv("penguins.csv")

# Create separate DataFrames for each species
adelie_data = df[df['species'] == 'Adelie']
chinstrap_data = df[df['species'] == 'Chinstrap']
gentoo_data = df[df['species'] == 'Gentoo']

# Create a scatterplot with all three species
plt.figure(figsize=(8, 6))

# Scatterplot for Adelie penguins
plt.scatter(adelie_data['bill_length_mm'], adelie_data['bill_depth_mm'], c='pink', label='Adelie', marker='o', s=100)

# Scatterplot for Chinstrap penguins
plt.scatter(chinstrap_data['bill_length_mm'], chinstrap_data['bill_depth_mm'], c='green', label='Chinstrap', marker='s', s=100)

# Scatterplot for Gentoo penguins
plt.scatter(gentoo_data['bill_length_mm'], gentoo_data['bill_depth_mm'], c='purple', label='Gentoo', marker='^', s=100)

plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Penguin Bill Length vs. Bill Depth')
plt.legend()

plt.show()





```

 **What do you observe from the graph?** (2 marks)

We can observe that Adelie species has greater bill depth and small bill length where as Gentoo has longer bill length and smaller bill depth. Chinstrap has moderate depth and length.

**Why did you choose this type of graph?** (2 marks)

Because it clearly differentiates between the three species. Different colours data points defining each species makes it easier to understand.

Suitable graph of the mean BMI (BMI not body_mass weight_in_kg/height(m2)) of female and male penguins. (Assume the height is the flipper length)

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("penguins.csv")

df['BMI'] = df['body_mass_g'] / ((df['flipper_length_mm'] / 1000) ** 2)  


mean_bmi_by_sex = df.groupby('sex')['BMI'].mean()


plt.figure(figsize=(8, 6))
mean_bmi_by_sex.plot(kind='bar', color=['green', 'purple'], alpha=0.7)
plt.xlabel('Sex')
plt.ylabel('Mean BMI')
plt.title('Mean BMI Penguins based on gender')
plt.xticks(rotation=0)
plt.show()
```

 **What do you observe from the graph?** (2 marks)

The bmi of female is smaller than male penguins. there isnt much difference but visible.

**Why did you choose this type of graph?** (2 marks)

Bar graph is the best graph to compare the data. We can easily compare male and female bmi here.*
